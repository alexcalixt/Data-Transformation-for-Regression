```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
%matplotlib inline
import scipy.stats
import pylab
```

# Import Data

```python
df = pd.read_csv('Car data.csv')
```

# EDA

```python
df.info()
```

```python
df.head()
```

```python
df.hist(figsize = (16,8))
```

# Other Transformation

### Selling Price

```python
df['t_Selling_Price'] = (df['Selling_Price']+1)**(1/14)

scipy.stats.probplot(df.t_Selling_Price, plot = pylab)
pylab.show()
plt.hist(df.t_Selling_Price)
plt.show()
```

### Kms_Driven

```python
df['t_Kms_Driven'] = (df['Kms_Driven']+1)**(1/14)
```

```python
sns.jointplot( x ='t_Kms_Driven',y ='t_Selling_Price', data = df, kind = 'reg')
```

```python
scipy.stats.probplot(df.t_Kms_Driven, plot = pylab)
pylab.show()
plt.hist(df.t_Kms_Driven)
plt.show()
```

### Present_Price

```python
df['t_Present_Price'] = (df['Present_Price']+1)**(1/14)
```

```python
sns.jointplot( x ='t_Present_Price',y ='t_Selling_Price', data = df, kind = 'reg')
```

```python
scipy.stats.probplot(df.t_Present_Price, plot = pylab)
pylab.show()
plt.hist(df.t_Present_Price)
plt.show()
```
