```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
%matplotlib inline
import scipy.stats
import pylab
```

# Import Data

```python
df = pd.read_csv('Car data.csv')
```

# EDA

```python
df.info()
```

# Reciprocal Transformation

### Selling Price

```python
df['inv_Selling_Price'] = 1/(df['Selling_Price']+1)
```

```python
scipy.stats.probplot(df.inv_Selling_Price, plot = pylab)
pylab.show()
plt.hist(df.inv_Selling_Price)
plt.show()
```

### Kms_Driven

```python
df['inv_Kms_Driven'] = 1/(df['Kms_Driven']+1)
```

```python
sns.jointplot( x ='inv_Kms_Driven',y ='inv_Selling_Price', data = df, kind = 'reg')
```

```python
scipy.stats.probplot(df.inv_Kms_Driven, plot = pylab)
pylab.show()
plt.hist(df.inv_Kms_Driven)
plt.show()
```

### Present Price

```python
df['inv_Present_Price'] = 1/(df['Present_Price']+1)
```

```python
sns.jointplot( x ='inv_Present_Price',y ='inv_Selling_Price', data = df, kind = 'reg')
```

```python
scipy.stats.probplot(df.inv_Present_Price, plot = pylab)
pylab.show()
plt.hist(df.inv_Present_Price)
plt.show()
```
