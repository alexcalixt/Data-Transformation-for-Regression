# Reciprocal Transformation

### Selling Price

```python
df['inv_Selling_Price'] = 1/(df['Selling_Price']+1)
```

### Kms_Driven

```python
sns.jointplot( x ='Kms_Driven',y ='Selling_Price', data = df, kind = 'reg')
```

```python
df['inv_Kms_Driven'] = 1/(df['Kms_Driven']+1)
```

```python
sns.jointplot( x ='inv_Kms_Driven',y ='inv_Selling_Price', data = df, kind = 'reg')
```

### Present Price

```python
sns.jointplot( x ='Present_Price',y ='Selling_Price', data = df, kind = 'reg')
```

```python
df['inv_Present_Price'] = 1/(df['Present_Price']+1)
```

```python
sns.jointplot( x ='inv_Present_Price',y ='inv_Selling_Price', data = df, kind = 'reg')
```
