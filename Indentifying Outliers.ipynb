```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
%matplotlib inline
```

# Import Data

```python
df = pd.read_csv('insurance.csv')
```

# Outlier Detention for Normal Variables

```python
def find_normal_oulier(df, var):
    
    upper_limit = df[var].mean() + 3 * df[var].std()
    lower_limit = df[var].mean() - 3 * df[var].std()
    
    print('Upper Limit:',upper_limit), print('Lower Limit:',lower_limit)
```

```python
plt.hist(df['bmi'])
plt.show()
sns.boxplot(y = df['bmi'])
```

```python
find_normal_oulier(df, 'bmi')
```

# Outlier Detection for Skewed Variables

```python
def find_skw_oulier(df, var, distance):
    IQR = df[var].quantile(0.75) - df[var].quantile(0.25) #Q3-Q1
    
    upper_limit = df[var].quantile(0.25) - (IQR * distance)
    lower_limit = df[var].quantile(0.75) + (IQR * distance)
    
    print('Upper Limit:',upper_limit), print('Lower Limit:',lower_limit)
```

```python
plt.hist(df['charges'])
plt.show()
sns.boxplot(y = df['charges'])
```

```python
find_skw_oulier(df, 'charges', 1.5)
```
