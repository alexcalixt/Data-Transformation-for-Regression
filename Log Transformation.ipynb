```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
%matplotlib inline
import scipy.stats
import pylab
```

# Import Data

```python
df = pd.read_csv('Car data.csv')
```

# EDA

```python
df.info()
```

```python
df.head()
```

```python
df.hist(figsize = (16,8))
```

# Log Transformation

### Selling Price

```python
df['log_Selling_Price'] = np.log(df['Selling_Price']+1)
```

```python
scipy.stats.probplot(df.log_Selling_Price, plot = pylab)
pylab.show()
plt.hist(df.log_Selling_Price)
plt.show()
```

### Kms_Driven

```python
df['log_Kms_Driven'] = np.log(df['Kms_Driven']+1)
```

```python
sns.jointplot( x ='log_Kms_Driven',y ='log_Selling_Price', data = df, kind = 'reg')
```

```python
scipy.stats.probplot(df.log_Kms_Driven, plot = pylab)
pylab.show()
plt.hist(df.log_Kms_Driven)
plt.show()
```

### Present_Price

```python
df['log_Present_Price'] = np.log(df['Present_Price']+1)
```

```python
sns.jointplot( x ='log_Present_Price',y ='log_Selling_Price', data = df, kind = 'reg')
```

```python
scipy.stats.probplot(df.log_Present_Price, plot = pylab)
pylab.show()
plt.hist(df.log_Present_Price)
plt.show()
```
