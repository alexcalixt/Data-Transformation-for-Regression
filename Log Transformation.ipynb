```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
%matplotlib inline
```

# Import Data

```python
df = pd.read_csv('Car data.csv')
```

# EDA

```python
df.info()
```

```python
df.head()
```

```python
df.hist(figsize = (16,8))
```

# Log Transformation

### Selling Price

```python
df['log_Selling_Price'] = np.log(df['Selling_Price']+1)
```

### Kms_Driven

```python
sns.jointplot( x ='Kms_Driven',y ='Selling_Price', data = df, kind = 'reg')
```

```python
df['log_Kms_Driven'] = np.log(df['Kms_Driven']+1)
```

```python
sns.jointplot( x ='log_Kms_Driven',y ='log_Selling_Price', data = df, kind = 'reg')
```

### Present_Price

```python
sns.jointplot( x ='Present_Price',y ='Selling_Price', data = df, kind = 'reg')
```

```python
df['log_Present_Price'] = np.log(df['Present_Price']+1)
```

```python
sns.jointplot( x ='log_Present_Price',y ='log_Selling_Price', data = df, kind = 'reg')
```
